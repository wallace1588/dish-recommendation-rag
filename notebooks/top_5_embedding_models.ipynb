```python
from langchain.embeddings import HuggingFaceEmbeddings
import numpy as np
from numpy.linalg import norm
import pandas as pd
from google.colab import drive
import csv
import time

drive.mount('/content/drive')
```

```python
vector_database_csv = '/content/drive/MyDrive/tcc_2024/files/dish_embedding.csv'
df_vector_database = pd.read_csv(vector_database_csv)

def list_str(A):
  A = A[1:-1]
  A = A.split(',')
  A = list(map(float, A))
  return np.array(A)

df_vector_database['mult_embedding_1'] = df_vector_database['mult_embedding_1'].apply(list_str)
df_vector_database['mult_embedding_2'] = df_vector_database['mult_embedding_2'].apply(list_str)
df_vector_database['mult_embedding_3'] = df_vector_database['mult_embedding_3'].apply(list_str)
df_vector_database['mult_embedding_4'] = df_vector_database['mult_embedding_4'].apply(list_str)
df_vector_database['port_embedding_1'] = df_vector_database['port_embedding_1'].apply(list_str)
df_vector_database['port_embedding_2'] = df_vector_database['port_embedding_2'].apply(list_str)
df_vector_database['port_embedding_3'] = df_vector_database['port_embedding_3'].apply(list_str)
df_vector_database['port_embedding_4'] = df_vector_database['port_embedding_4'].apply(list_str)

df_vector_database
```

```python
def combine_columns_embed(row):
  text_row = row['name'] + row['details']
  return text_row

df_vector_database['check'] = df_vector_database.apply(combine_columns_embed, axis=1)

df_vector_database = df_vector_database.drop_duplicates(subset=['check'])

df_vector_database
```

```python
class llm:

  def __init__(self, n_llm, n_vector_index, name):
    self.embeddings = HuggingFaceEmbeddings(
        model_name=n_llm
    )
    self.vector_index = n_vector_index
    self.name = name


  def get_similarity(self, query, k, df):
    df_res = df.copy()
    B = np.array(self.embeddings.embed_query(query))
    A = df_res[self.vector_index]

    df_res['score'] = df_res[self.vector_index].apply(lambda a : np.dot(a,B)/(norm(a)*norm(B)))

    vector_similarity = df_res.sort_values( by="score", ascending=False)

    return vector_similarity[['name','details', 'price']][0:k].copy()
```

```python
class paraphrase_multilingual_MiniLM_L12_v2(llm):
  n_llm = 'sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2'
  n_vector_index = 'mult_embedding_1'
  name = 'paraphrase-multilingual-MiniLM-L12-v2'

  def __init__(self):
        super().__init__(self.n_llm, self.n_vector_index, self.name)

class distiluse_base_multilingual_cased_v2(llm):
  n_llm = 'sentence-transformers/distiluse-base-multilingual-cased-v2'
  n_vector_index = 'mult_embedding_2'
  name = 'distiluse-base-multilingual-cased-v2'

  def __init__(self):
        super().__init__(self.n_llm, self.n_vector_index, self.name)

class paraphrase_multilingual_mpnet_base_v2(llm):
  n_llm = 'sentence-transformers/paraphrase-multilingual-mpnet-base-v2'
  n_vector_index = 'mult_embedding_3'
  name = 'paraphrase-multilingual-mpnet-base-v2'

  def __init__(self):
        super().__init__(self.n_llm, self.n_vector_index, self.name)

class multilingual_e5_small(llm):
  n_llm = 'intfloat/multilingual-e5-small'
  n_vector_index = 'mult_embedding_4'
  name = 'multilingual-e5-small'

  def __init__(self):
        super().__init__(self.n_llm, self.n_vector_index, self.name)

class bert_large_portuguese_cased_legal_mlm_nli_sts_v1(llm):
  n_llm = 'stjiris/bert-large-portuguese-cased-legal-mlm-nli-sts-v1'
  n_vector_index = 'port_embedding_1'
  name = 'bert-large-portuguese-cased-legal-mlm-nli-sts-v1'

  def __init__(self):
        super().__init__(self.n_llm, self.n_vector_index, self.name)

class Legal_BERTimbau_sts_base(llm):
  n_llm = 'rufimelo/Legal-BERTimbau-sts-base'
  n_vector_index = 'port_embedding_2'
  name = 'Legal-BERTimbau-sts-base'

  def __init__(self):
        super().__init__(self.n_llm, self.n_vector_index, self.name)

class msmarco_distilbert_base_tas_mmarco_pt_300k(llm):
  n_llm = 'mpjan/msmarco-distilbert-base-tas-b-mmarco-pt-300k'
  n_vector_index = 'port_embedding_3'
  name = 'msmarco-distilbert-base-tas-b-mmarco-pt-300k'

  def __init__(self):
        super().__init__(self.n_llm, self.n_vector_index, self.name)

class ptbr_similarity_e5_small(llm):
  n_llm = 'jmbrito/ptbr-similarity-e5-small'
  n_vector_index = 'port_embedding_4'
  name = 'ptbr-similarity-e5-small'

  def __init__(self):
        super().__init__(self.n_llm, self.n_vector_index, self.name)
```

```python
query1 = "jantar vegano"
query2 = "comida tipica brasileira"
query3 = "algo refrescante"
query4 = "opção vegana de almoço"
query5 = "sobremesa de frutas"
query6 = "hamburguer com batata frita mais refrigerante"
query7 = "pizza de queijo"
query8 = "comida japonesa"
query9 = "lanche com suco"
query10 = "almoço com feijão e carne, sem verdura"

querys = [query1, query2, query3, query4, query5, query6, query7, query8, query9, query10]
```

```python
model1 = paraphrase_multilingual_MiniLM_L12_v2()
model2 = distiluse_base_multilingual_cased_v2()
model3 = paraphrase_multilingual_mpnet_base_v2()
model4 = multilingual_e5_small()
model5 = ptbr_similarity_e5_small()
model6 = bert_large_portuguese_cased_legal_mlm_nli_sts_v1()
model7 = Legal_BERTimbau_sts_base()
model8 = msmarco_distilbert_base_tas_mmarco_pt_300k()

models = [model1, model2, model3, model4, model5, model6, model7, model8]
```

```python
result = []
with open('/content/drive/MyDrive/tcc_2024/files/outputs_columns.csv', 'w', newline='') as file:
    writer = csv.writer(file)

    k = 5

    for query in querys:
      lines = []
      times = []
      writer.writerow(['',query])
      add_row = []
      for model in models:
        add_row = add_row + [model.name, 'Avaliações', '']
      writer.writerow([''] + add_row)
      for model in models:
        ex_time = time.time()
        result = model.get_similarity(query, k, df_vector_database)
        ex_time = time.time() - ex_time
        times.append(ex_time)
        rows = len(result)
        columns = []
        for i in range(rows):
          columns.append(str(result.iloc[i,0]) + ': ' + str(result.iloc[i,1]))
        lines.append(columns)

      for i in range(k):
        add_row = []
        for j in range(len(lines)):
          add_row = add_row + [lines[j][i],'0','']
        writer.writerow([i+1] + add_row)

      writer.writerow([''])

      add_row = []
      for t in times:
        add_row = add_row + ["Time execution: " + str(t), '','']

      writer.writerow([''] + add_row)
      writer.writerow([''])
      writer.writerow([''])


result
```
