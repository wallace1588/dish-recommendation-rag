```python
from langchain.embeddings import HuggingFaceEmbeddings
from google.colab import drive
import pandas as pd
import csv
import sys
from IPython.display import clear_output
```

```python
embedding_model_1 = "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
embedding_model_2 = "sentence-transformers/distiluse-base-multilingual-cased-v2"
embedding_model_3 = "sentence-transformers/paraphrase-multilingual-mpnet-base-v2"
embedding_model_4 = "intfloat/multilingual-e5-small"
embedding_model_5 = "stjiris/bert-large-portuguese-cased-legal-mlm-nli-sts-v1"
embedding_model_6 = "rufimelo/Legal-BERTimbau-sts-base"
embedding_model_7 = "mpjan/msmarco-distilbert-base-tas-b-mmarco-pt-300k"
embedding_model_8 = "jmbrito/ptbr-similarity-e5-small"
```

```python
drive.mount('/content/drive')

csv.field_size_limit(sys.maxsize)

df_dish = pd.read_csv('/content/drive/MyDrive/tcc_2024/files/dish.csv')
df_dish = df_dish[['_id', 'name', 'details', 'price']]

df_dish
```

```python
df_dish = df_dish[pd.notna(df_dish['details'])]
df_dish
```

```python
# Define Embeddings using HF
embeddings_1 = HuggingFaceEmbeddings(model_name=embedding_model_1)
embeddings_2 = HuggingFaceEmbeddings(model_name=embedding_model_2)
embeddings_3 = HuggingFaceEmbeddings(model_name=embedding_model_3)
embeddings_4 = HuggingFaceEmbeddings(model_name=embedding_model_4)
embeddings_5 = HuggingFaceEmbeddings(model_name=embedding_model_5)
embeddings_6 = HuggingFaceEmbeddings(model_name=embedding_model_6)
embeddings_7 = HuggingFaceEmbeddings(model_name=embedding_model_7)
embeddings_8 = HuggingFaceEmbeddings(model_name=embedding_model_8)
```

```python
embeddings_list_1 = []
embeddings_list_2 = []
embeddings_list_3 = []
embeddings_list_4 = []
embeddings_list_5 = []
embeddings_list_6 = []
embeddings_list_7 = []
embeddings_list_8 = []

rows_df = len(df_dish)
list_column = df_dish.columns.values.tolist()

for i in range(rows_df):

  if i % 50 == 0:
    clear_output()
    print(str(rows_df) + '/' + str(i))

  txt_emb = str(df_dish.iloc[i,1] + ': ' + df_dish.iloc[i,2])

  embeddings_list_1.append(embeddings_1.embed_query(txt_emb))
  embeddings_list_2.append(embeddings_2.embed_query(txt_emb))
  embeddings_list_3.append(embeddings_3.embed_query(txt_emb))
  embeddings_list_4.append(embeddings_4.embed_query(txt_emb))
  embeddings_list_5.append(embeddings_5.embed_query(txt_emb))
  embeddings_list_6.append(embeddings_6.embed_query(txt_emb))
  embeddings_list_7.append(embeddings_7.embed_query(txt_emb))
  embeddings_list_8.append(embeddings_8.embed_query(txt_emb))


df_dish.insert(len(list_column)  , 'mult_embedding_1', embeddings_list_1, True)
df_dish.insert(len(list_column)+1, 'mult_embedding_2', embeddings_list_2, True)
df_dish.insert(len(list_column)+2, 'mult_embedding_3', embeddings_list_3, True)
df_dish.insert(len(list_column)+3, 'mult_embedding_4', embeddings_list_4, True)
df_dish.insert(len(list_column)+4, 'port_embedding_1', embeddings_list_5, True)
df_dish.insert(len(list_column)+5, 'port_embedding_2', embeddings_list_6, True)
df_dish.insert(len(list_column)+6, 'port_embedding_3', embeddings_list_7, True)
df_dish.insert(len(list_column)+7, 'port_embedding_4', embeddings_list_8, True)

clear_output()
print(str(rows_df) + '/' + str(rows_df))

df_dish
```

```python
df_dish.to_csv('/content/drive/MyDrive/tcc_2024/files/dish_embedding.csv', index= False)

df_dish
```
